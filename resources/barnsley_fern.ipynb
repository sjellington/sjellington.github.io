```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fn1(x,y):
    return (0.0*x, 0.16*y)
def fn2(x,y):
    return (0.85*x + 0.04*y, -0.04*x + 0.85*y + 1.6)
def fn3(x,y):
    return (0.2*x - 0.26*y, 0.23*x + 0.22*y + 1.6)
def fn4(x,y):
    return (-0.15*x + 0.28*y, 0.26*x + 0.24*y + 0.44)
fns = [fn1, fn2, fn3, fn4]
```

```python
points = 75000
x, y = 0, 0
x_list = []
y_list = []
for i in range(points):
    function = np.random.choice(fns, p=[0.01, 0.85, 0.07, 0.07])
    x, y = function(x, y)
    x_list.append(x)
    y_list.append(y)
```

```python
plt.scatter(x_list, y_list, s=0.1, color='green')
```
