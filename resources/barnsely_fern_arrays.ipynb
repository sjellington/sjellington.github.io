```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
def fn1(x,y):
    return (0.0*x, 0.16*y)
def fn2(x,y):
    return (0.85*x + 0.04*y, -0.04*x + 0.85*y + 1.6)
def fn3(x,y):
    return (0.2*x - 0.26*y, 0.23*x + 0.22*y + 1.6)
def fn4(x,y):
    return (-0.15*x + 0.28*y, 0.26*x + 0.24*y + 0.44)
fns = [fn1, fn2, fn3, fn4]
```

```python
def func1(x,y):
    return (0.0*x, 0.25*y - 0.4)
def func2(x,y):
    return (0.95*x + 0.005*y -0.002, -0.005*x + 0.93*y + 0.5)
def func3(x,y):
    return (0.035*x - 0.2*y -0.09, 0.16*x + 0.04*y + 0.02)
def func4(x,y):
    return (-0.04*x + 0.2*y + 0.083, 0.16*x + 0.04*y + 0.12)
fns = [func1, func2, func3, func4]
```

```python
points = 100000
width, height = 500,500
fern_image = np.zeros((width, height))
x, y = 0, 0

for i in range(points):
    function = np.random.choice(fns, p=[0.02, 0.84, 0.07, 0.07])
    x, y = function(x,y)
    ix, iy = int(width / 2 + x * width / 10), int(y * height / 12)
    fern_image[iy,ix] = 1

plt.imshow(fern_image[::-1,:], cmap=cm.Greens)
plt.show
```
